# 不要な列や行の削除

```python
columns = df.columns
columns
```

```python
# .1や.2などが絡む名に付いているものを削除
def col_filter(c):
    if('.1' in c):
        return False
    elif('.2' in c):
        return False
    else:
        return True
filtered = list(filter(col_filter, columns))
filtered
```

```python
df[filtered]
```

```python
# 0行目はNaNなのでスキップして1行目から読み込む
df[filtered][1:]
```

```python
df = df[filtered][1:]
df.head()
```

# カラム名変更

```python
df.columns
```

```python
# カラム属性に代入する方法
df.columns = [
    '年月日',
    '平均気温',
    '最高気温',
    '最低気温',
    '降水量の合計',
    '最深積雪',
    '平均雲量',
    '平均蒸気圧',
    '平均風速',
    '日照時間'
]
df.head()
```
