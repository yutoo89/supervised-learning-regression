# Seleniumで自動ログイン
詳細はwebスクレイピングの動画

````python
!pip install -U selenium
````

chrome（ブラウザ）のバージョンと、chrom-driver（ブラウザを操作するツール）のバージョンが違う、というエラーがよく出る。
そのエラーを防ぐためにwebdriver_managerを使う

````python
!pip install webdriver_manager
````

````python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
````

## Chromeブラウザのインストール方法
動画では省略されていたため、自分でインストールした際のメモ

```bash
wget https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb;
apt install  --fix-broken ./google-chrome-stable_current_amd64.deb
```

上記でインストールできるらしい。

```bash
E: ロック /var/lib/dpkg/lock-frontend が取得できませんでした - open (11: リソースが一時的に利用できません)
```

このエラーが出た場合は対象ファイルを削除すればよさそう。

```bash
rm /var/lib/dpkg/lock-frontend
```

````python
from selenium.webdriver.chrome.options import Options
# ブラウザの立ち上げ
# ChromeDriverManagerでchromeのバージョンに合ったdriverをインストールする
driver = ChromeDriverManager().install()
# driver
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome(driver, options=options)
````
